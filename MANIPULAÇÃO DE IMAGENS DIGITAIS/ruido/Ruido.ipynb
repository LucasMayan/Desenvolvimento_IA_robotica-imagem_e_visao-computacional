Filtros

```python
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np
```

```python
img = cv.imread('ruido.png')

plt.imshow(img)
```

```python
kernal = np.ones((5,5),np.float32)/25
dst = cv.filter2D(img, -1, kernal)
plt.imshow(dst)
```

```python
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.subplot(122),plt.imshow(dst),plt.title('Media 1x')

plt.show()
plt.clf()
```

```python
blur = cv.blur(img, (5,5))
plt.imshow(blur)
```

Morfologia

```python
img1 = cv.imread('assinatura.png', 0)
struct = np.ones((5,5), np.uint8)
plt.imshow(img1)
```

```python
img_erosao = cv.erode(img1, struct, iterations=3)
plt.imshow(img_erosao)
```

```python
img_dilatacao = cv.dilate(img1, struct, iterations= 1)
plt.imshow(img_dilatacao)
```

```python
img_erosao2 = cv.erode(img_dilatacao, struct, iterations=3)
plt.imshow(img_erosao2)
```
