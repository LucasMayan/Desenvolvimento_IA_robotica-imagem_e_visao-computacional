```python
def extract_data(arquivo):
    imgRet = cv.imread(arquivo, 0)
    imgRet = cv.cvtColor(imgRet, cv.COLOR_RGB2BGR)
    imgRet = imgRet.ravel()
    imgMedia = np.mean(imgRet)

    return imgMedia
```

```python
extract_data("cena1.png")
```

```python
lista_arquivos = listdir("imgs")
lista_arquivos.sort()

lista_Medias = list()

for nome_arquivo in lista_arquivos:
    print(extract_data('imgs/' + nome_arquivo))
    media = extract_data('imgs/' + nome_arquivo)
    lista_Medias.append(media)

plt.plot(lista_Medias)
```
