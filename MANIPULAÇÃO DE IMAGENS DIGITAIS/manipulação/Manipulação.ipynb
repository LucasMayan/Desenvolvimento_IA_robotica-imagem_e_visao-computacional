```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Carrega a imagem
imagem = cv2.imread("profile_pic.jpg")
plt.imshow(imagem)
```

```python
imagem = cv2.cvtColor(imagem, cv2.COLOR_RGB2BGR)
plt.imshow(imagem)
```

```python
def embacar(img, tamanho):
    blurImg = cv2.blur(imagem,(tamanho,tamanho))
    return blurImg
```

```python
plt.imshow(embacar(imagem, 3))
```

```python
aguca = np.array([
    [ -1, -1, -1],
    [1, 0, -1],
    [1, 1, 0]
])
```

```python
detect = np.array([
    [ 0, -1, -1],
    [1, 0, -1],
    [1, 1, 0]
])
```

```python
sol = np.array([
    [ 0.272, 0.534, 0.131],
    [0.349, 0.686, 0.168],
    [0.393, 0.769, 0.189]
])
```

```python
def filtro_conv(img, kernal):
    return cv2.filter2D(img, -1, kernal)
```

```python
plt.imshow(filtro_conv(imagem, aguca))
```

```python
plt.imshow(filtro_conv(imagem, detect))
```

```python
plt.imshow(filtro_conv(imagem, sol))
```
