```python
import cv2
import matplotlib.pyplot as plt
```

```python
imagem = cv2.imread("profile_pic.jpg")
plt.imshow(imagem)
```

```python
imagem = cv2.cvtColor(imagem, cv2.COLOR_RGB2BGR)
plt.imshow(imagem)
```

```python
plt.hist(imagem.ravel())
```

```python
imgRavel = imagem.ravel()
```

```python
imagempb = cv2.imread("profile_pic.jpg", 0)
plt.imshow(imagempb)
```

```python
imgRavelpb = imagempb.ravel()
plt.hist(imgRavelpb)
```

```python
equ = cv2.equalizeHist(imagempb)
plt.imshow(equ)
```

```python
equRavel = equ.ravel()
plt.hist(equRavel)
```
